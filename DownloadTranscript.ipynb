# 🤖 Generate Debate transcription using Whisper

**Important**: To try this, verify that the GPU is enabled (*Runtime* > *Change runtime time* > Seleccionar T4 GPU como *Hardware accelerator*)


```python
!pip install git+https://github.com/m-bain/whisperx.git --upgrade
!pip install pytube

from pytube import YouTube
import whisperx
import gc
import os
```

```python
# Save audio file in '/content'
video_url = "https://www.youtube.com/watch?v=0PfphtXa6lQ"
yt = YouTube(video_url)
audio_stream = yt.streams.filter(only_audio=True, file_extension='mp4').first()
output_path = "/content"
audio_stream.download(output_path, filename="audio.mp3")

print(f"Audio downloaded to {output_path}")
```

Transcript generation with WhisperX


```python
audio_file = "/content/audio.mp3"

device = "cuda"
batch_size = 16
compute_type = "float16"

model = whisperx.load_model("large-v2", device, compute_type=compute_type)
audio = whisperx.load_audio(audio_file)
result = model.transcribe(audio, batch_size=batch_size)
print(result["segments"])

model_a, metadata = whisperx.load_align_model(language_code=result["language"], device=device)
result = whisperx.align(result["segments"], model_a, metadata, audio, device, return_char_alignments=False)
print(result["segments"])
```

Saving transcription in /content/transcript.txt

```python
folder_path = "/content"
file_name = "transcript.txt"

file_path = os.path.join(folder_path, file_name)

transcription_text = "\n".join([segment['text'] for segment in result["segments"]])

try:
    with open(file_path, "w") as f:
        f.write(transcription_text)
except Exception as e:
    print(f"Error: {e}")
```
